```python
import numpy as np                       # Importing numpy library
import matplotlib.pyplot as plt          # Importing matplotlib library for plotting
from astropy import units as u           # Importing Astropy units for handling physical quantities with units
from astropy.constants import L_sun      # Importing specific constants from Astropy constants module
```

```python
print(L_sun)   # Let's check the luminosity of Sun and it's unit
```

```python
# Known values of flux and luminosity for Betelgeuse
flux = 3.27e-8 * u.W / u.m**2
luminosity = 1.2e4 * L_sun
print('Flux = ', flux)
print('Luminosity = ', luminosity)
```

```python
# Now you have everything in the correct unit so
# Compute distance (D) using the equation F = L / (4 * pi * D^2)
# Note: np.pi gives you: 3.141592653589793 (you can print it to make sure!)
distance_sq = ((luminosity / (4 * np.pi * flux))) #.to(u.pc)
distance = np.sqrt(distance_sq)
print(distance)
```

```python
# Convert the distance unit from m to pc
print('Distance to Betelgeuse = ', distance.to(u.pc))
```

## Example 2.2: Parallax Distance

Write a function to calculate the distance to a star using the parallax method. Note that you have the measurements for the angle $\theta$.
And you need to use the following formula:

$d\,[\rm pc] = \dfrac{1\, [\rm AU]}{\theta\,[\rm arcsecond]}$



units:

$1\, [\rm arcsec] = \dfrac{1}{3600}\, [deg]$

$1\, \rm [pc] = 3.26 \, [\rm ly]$

$1\,\rm [AU] = 149.6 \times 10^{9}\,[m] $



```python

def compute_parallax_distance(parallax_angle_arcsec):
    # Constants
    astronomical_unit = u.au  # 1 astronomical unit in meters

    # Convert parallax angle to radians
    parallax_angle_rad = parallax_angle_arcsec * u.arcsec.to(u.rad)

    # Compute the parallax distance using d = 1[AU] / tan(theta)
    parallax_distance = astronomical_unit / np.tan(parallax_angle_rad)

    # Return the result in parsecs
    return parallax_distance.to(u.pc)


parallax_angle_arcsec = 0.2  # Change this value to any parallax angle you want for the exercise
distance = compute_parallax_distance(parallax_angle_arcsec)
print(f"Parallax distance: {distance:.2f}")
```

# Example 3.1: Comoving Distance ($D_C$)

Compute the Comoving Distance for a range of redshift in some fixed cosmology. Then check how much this distance could vary by changing the cosmological model of our Universe.

# Reminder:

Comoving Distance:

* Distance in coordinates that do expand with expanding universe.

* Measures the current physical separation between two objects in the universe, accounting for the expansion of space over time.


$H^{2} = H^2_0\bigg(\Omega_m(1+z)^3 + \Omega_K(1+z)^2 + \Omega_{\Lambda} \bigg)$

$D_C = c\int_0^z \dfrac{dz^{\prime}}{H(z^{\prime})}$  

# Note:
When we assume a ***Flat Universe***:
* $\Omega_K = 0$
* $\Omega_{\Lambda} = 1 - \Omega_m$

```python
# Import necessary libraries
from astropy.cosmology import FlatLambdaCDM # This library gives us a flat LCDM model.
```

```python
# Define the cosmology model
omega_m = 0.3  # Matter density parameter
H = 70  # Hubble constant (km/s/Mpc)
cosmo = FlatLambdaCDM(H0=H, Om0=omega_m)

# Define the redshift range to calculate the distance
z = np.linspace(0, 10, 100)

# Comoving distance
comoving_dist = cosmo.comoving_distance(z)  # Comoving distance

# Plot the results:
plt.plot(z, comoving_dist, label='Comoving Distance')
plt.xlabel(r'$z$')
plt.ylabel(r'$D_C$')

plt.legend()
```

```python
# Compare different cosmological models
# keep the H0 value fixed and change the amount of matter in the Universe
cosmo1 = FlatLambdaCDM(H0=70, Om0=0.05)
cosmo2 = FlatLambdaCDM(H0=70, Om0=0.3)
cosmo3 = FlatLambdaCDM(H0=70, Om0=1)

# Compute the comoving distance for each case
comoving_dist1 = cosmo1.comoving_distance(z)  # Comoving distance
comoving_dist2 = cosmo2.comoving_distance(z)  # Comoving distance
comoving_dist3 = cosmo3.comoving_distance(z)  # Comoving distance

# Plot the results to see the comparison
plt.plot(z, comoving_dist1, label='$\Omega_m = 0.05, \, \Omega_{\Lambda} = 0.95$')
plt.plot(z, comoving_dist2, label='$\Omega_m = 0.3, \, \Omega_{\Lambda} = 0.7$')
plt.plot(z, comoving_dist3, label='$\Omega_m = 1, \, \Omega_{\Lambda} = 0$')

plt.text(0,12500, '$H_0 = 70\, [km/s/Mpc]$')
plt.title('Flat $\Lambda$CDM')
plt.xlabel(r'$z$')
plt.ylabel(r'$D_C$')
plt.legend()
```

```python
# Now keep the amount of matter in the Universe fixed and change the change the value of H0
# to see how much this value affecting the distances to the objects

cosmo4 = FlatLambdaCDM(H0=40, Om0=0.3)
cosmo5 = FlatLambdaCDM(H0=70, Om0=0.3)
cosmo6 = FlatLambdaCDM(H0=100, Om0=0.3)

# Compute the comoving distance for each case
comoving_dist4 = cosmo4.comoving_distance(z)  # Comoving distance
comoving_dist5 = cosmo5.comoving_distance(z)  # Comoving distance
comoving_dist6 = cosmo6.comoving_distance(z)  # Comoving distance

# Plot the results to see the differences
plt.plot(z, comoving_dist4, label='$H_0 = 40$')
plt.plot(z, comoving_dist5, label='$H_0 = 70$')
plt.plot(z, comoving_dist6, label='$H_0 = 100$')

plt.text(0,12500, '$\Omega_m = 0.3.\,\Omega_{\Lambda}=0.7$')
plt.title('Flat $\Lambda$CDM')
plt.xlabel(r'$z$')
plt.ylabel(r'$D_C$')
plt.legend()
```

# Reminder:

Angular Diameter Distance:

The ratio of an object's physical size to its angular size as observed from Earth. It represents the physical distance between the observer and the object, taking into account the effects of the expanding universe.

$H^{2} = H^2_0\bigg(\Omega_m(1+z)^3 + \Omega_K(1+z)^2 + \Omega_{\Lambda} \bigg)$

$D_A = \dfrac{x}{\theta}$

$D_A = D_C/(1+z)$

# Note:
When we assume a ***Flat Universe***:
* $\Omega_K = 0$
* $\Omega_{\Lambda} = 1 - \Omega_m$

```python
# Define the cosmology model
omega_m = 0.3  # Matter density parameter
H = 70  # Hubble constant (km/s/Mpc)
cosmo = FlatLambdaCDM(H0=H, Om0=omega_m)

z = np.linspace(0, 10, 100)  # Range of redshift values

angular_diam_dist = cosmo.angular_diameter_distance(z)  # Angular diameter distance

# plot
plt.plot(z, angular_diam_dist, label='Angular Diameter Distance')
plt.xlabel(r'$z$')
plt.ylabel(r'$D_A$')

plt.legend()
```

```python
# Compare different cosmological models
# keep the H0 value fixed and change the amount of matter in the Universe
cosmo1 = FlatLambdaCDM(H0=70, Om0=0.05)
cosmo2 = FlatLambdaCDM(H0=70, Om0=0.3)
cosmo3 = FlatLambdaCDM(H0=70, Om0=1)

# Compute the luminosity distance for each case
angular_diam_dist1 = cosmo1.angular_diameter_distance(z)  # Angular diameter distance
angular_diam_dist2 = cosmo2.angular_diameter_distance(z)  # Angular diameter distance
angular_diam_dist3 = cosmo3.angular_diameter_distance(z)  # Angular diameter distance

# Plot the results to see the differences
#plt.plot(z, comoving_dist1/(1+z))
plt.plot(z, angular_diam_dist1, label='$\Omega_m = 0.05, \, \Omega_{\Lambda} = 0.95$')
plt.plot(z, angular_diam_dist2, label='$\Omega_m = 0.3, \, \Omega_{\Lambda} = 0.7$')
plt.plot(z, angular_diam_dist3, label='$\Omega_m = 1, \, \Omega_{\Lambda} = 0$')


plt.text(6,2400, '$H_0 = 70\, [km/s/Mpc]$')
plt.title('Flat $\Lambda$CDM')
plt.xlabel(r'$z$')
plt.ylabel(r'$D_A$')
plt.legend()
```
